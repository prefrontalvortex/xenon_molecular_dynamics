```python
def norm(x, axis=None):
    return np.sqrt(np.sum(np.array(x)**2, axis=axis))

def mark_spot(N, y0=200, y1=400, c='r'):
    plt.plot((N, N), (y0, y1), c=c)
```

```python
def histo(ary, nbins=32, vmin=0, vmax=1023):
    bins = [0,]*nbins
    vrange = vmax - vmin
    vstep = vrange / float(nbins)
    for item in ary:
        binval = int((item - vmin) // vstep)-1
        binval = nbins - 1 if binval >= nbins else binval
        bins[binval] = bins[binval] +1
    return bins
```

```python
histo(range(2,32,4), nbins=16, vmax=32)
```

```python
nbins=10
vmax=np.amax(spd)
print(vmax)
(spd*nbins/vmax).hist(bins=nbins)
plt.plot(histo(spd.as_matrix(), nbins=nbins+2, vmax=vmax), 'r.-')
```

```python
"""
int *buildHist(int bins, double min, double max, int n, double *data){
   double *hist=malloc(bins*sizeof(int));
   if (hist == NULL) return hist;
   for (int i=0; i<n; ++i){
      int bin=int( (data[i]-min)/((max-min)/(bins)) );
      if ( (bin>=0) && (bin<n) ) hist[bin]++;
   }
   return hist;
}
"""

def buildHist(bins, amin, amax, data):
    n = len(data)
    hist=np.zeros(bins+1)
    for i in range(n):
        abin=int( (data[i]-amin)/((amax-amin)/(bins)) )
        if ( (abin>=0) and (abin<n) ): hist[abin]+=1
   
    return hist;
```

```python
(spd*(nbins)/vmax).hist(bins=nbins)
plt.plot(buildHist(bins=nbins, amin=0, amax=vmax, data=spd.as_matrix()), 'r.-')
```
